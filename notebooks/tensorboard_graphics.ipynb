```python
import json
import matplotlib.pyplot as plt
```

```python
json_file_path = "2d_convex.json"

with open(json_file_path, "r", encoding="utf-8") as f:
    data = json.load(f)

steps = [point[1] for point in data]
rewards = [point[2] for point in data]

plt.figure(figsize=(12, 8))
plt.plot(steps, rewards, label="Средняя награда")

plt.xlabel("Шаги обучения", fontsize=12)
plt.ylabel("Средняя награда за эпизод", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```
